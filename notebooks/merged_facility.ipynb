```python
import pandas as pd
import os

# Set path to processed folder
processed_dir = r"C:\Users\Students\Desktop\flax_projects\SDI_Tanzania_Healthcare_Delays\data\processed"

facility_full = pd.read_csv(os.path.join(processed_dir, 'facility_full.csv'))
```

```python
# Impute with mode
for col in ['facility_level', 'has_sharps', 'vaccine_store2']:
    facility_full[col].fillna(facility_full[col].mode()[0])
```

```python
facility_full.isnull().sum()
```

```python
for col in ['facility_level', 'has_sharps', 'vaccine_store2']:
    print(col, facility_full[col].unique())
```

```python
for col in ['facility_level', 'has_sharps', 'vaccine_store2']:
    mode_value = facility_full[col].mode()[0]   # get the most frequent value
    facility_full[col] = facility_full[col].fillna(mode_value)

# Verify
print(facility_full[['facility_level', 'has_sharps', 'vaccine_store2']].isnull().sum())
```

```python
# -----------------------------
# Feature engineering
# -----------------------------
# Medical staff ratio
facility_full['med_staff_ratio'] = facility_full['num_med'] / facility_full['num_staff']
```

```python
# Proportion of staff present
facility_full['prop_present'] = facility_full['provider_present1']
```

```python
# Facility readiness score
facility_full['readiness_score'] = (
    facility_full['has_sharps'] +
    facility_full['vaccine_store2'] +
    facility_full['transport_central_a'] +
    facility_full['transport_central_hr'] +
    facility_full['transport_central_mn']
)
```
